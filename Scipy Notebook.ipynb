```python
b = special.exp2(3) #2^3
print(b)
```

# Double Integration 

```python
e = lambda x,y : x*y**2
f = lambda x: 1
g = lambda x: -1
integrate.dblquad(e,0,2,f,g)
```

# Fourier Transformations

Fourier Analysis is a Method that deals with Expressing a Function  as a sum of Periodic Components and Recovering the Signal from those Components

```python
from scipy.fftpack import fft,ifft
import numpy as np

x = np.array([1,2,3,4])
y = fft(x)

print(x)
print("\n\n",y)
```

```python
#To Find Inverse 

y = ifft(x)
print(y)
```

# Linear Algebra 

```python
from scipy import linalg
a = np.array([[1,2],[3,4]])
b = linalg.inv(a) #inverse of a matrix
print(b)

```

# Interpolated Function

Interpolation refers to Constructing new data points within a setting  of Known data points 

```python
import matplotlib.pyplot as plt
from scipy import interpolate

x = np.arange(5,20)
y = np.exp(x/3.0)

f = interpolate.interp1d(x,y)

x1 = np.arange(6,12)
y1 = f(x1) #use interpolation function returned by 'interp1d'

plt.plot(x,y,'o',x1,y1,'--')
plt.show()
```
